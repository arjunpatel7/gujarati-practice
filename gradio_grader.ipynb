```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
from pathlib import Path

LABELS = Path('class_names.txt').read_text().splitlines()

model = tf.keras.models.load_model("models/scratch_guju")
```

```python
#def predict(img):
    
    #create a batch of one
#    print(img)
#    img_array = np.zeros((224, 224))
#    print(img_array)
#    img_array[:img.shape[0],:img.shape[1]] = img
#    print(img_array.shape)
#    img_array = tf.image.grayscale_to_rgb(img_array)
#    img_array = tf.expand_dims(img_array, 0)
    
#    predictions = model.predict(img_array, batch_size=None)
#    confidences = {LABELS[i]: pred[i] for i in predictions}
#    return confidences





def predict(img):
    
    #create a batch of one
    img_array = np.zeros((28, 28))
    img_array[:img.shape[0],:img.shape[1]] = img
    img_array = tf.expand_dims(img_array, 0)
    
    predictions = model.predict(img_array, batch_size=None)[0].tolist()
    print(predictions)
    
    confidences = {}
    for i in range(0, len(predictions)):
        confidences[LABELS[i]] = predictions[i]
    print(confidences)
    #confidences = {LABELS[i]: predictions[i] for i in range(0, len(predictions))}
    return confidences
```

```python
gr.Interface(fn=predict, 
             inputs="sketchpad",
             outputs="label",
             live=True).launch()
```
